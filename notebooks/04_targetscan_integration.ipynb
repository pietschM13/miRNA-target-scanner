## Day 1: Loading and inspecting TargetScan data

```python
import pandas as pd

df_mirna = pd.read_csv('../data/human_mature_miRNAs_features.csv')
df_mirna.head()
```

```python
df_targets = pd.read_csv('../data/Predicted_Targets_Context_Scores.default_predictions.txt', sep='\t')
df_targets.head()
```

```python
df_targets.describe()
```

```python
# Create df with select columns

df_targetscan = df_targets[['miRNA', 'context++ score', 'context++ score percentile']]
df_targetscan = df_targetscan.rename(columns={'context++ score': 'context_score', 'context++ score percentile': 'context_percentile'})
df_targetscan.head()
```

```python
# FIlter for human miRNAs 

df_targetscan = df_targetscan[df_targetscan['miRNA'].str.startswith('hsa-')]
df_targetscan.describe()
```

```python
# Control for unique miRNAs

df_targetscan["miRNA"].nunique()
```

```python
df_targetscan.head(10)
```

```python
import re

# Find miRNA names that don't follow standard pattern
# Standard pattern: ends with '-number', '-3p', '-5p', or '-number+letter' (e.g., -302e, -23c)
standard_pattern = r'-(3p|5p|\d+[a-z]?)$' # define miRNA pattern using regex. 

outliers = [] #create list of potential outliers
for mirna in df_targetscan['miRNA'].unique(): #loop through miRNAs in df_targetscan 
    if not re.search(standard_pattern, mirna): # condition the loop to only include miRNAs that do not follow the defined standard pattern (i.e. outliers)
        outliers.append(mirna)  # add outliers to outliers list

print(f"Found {len(outliers)} outlier miRNA names:") # print outliers count
for outlier in sorted(outliers): # loop through outliers list and sort 
    print(f"  {outlier}") # print miRNA outliers
```

```python
# Replacing suffixes from miRNA outliers with 'no characters'

df_targetscan['miRNA'] = df_targetscan['miRNA'].str.replace(r'\.\d+$', '', regex=True)
df_targetscan['miRNA'].value_counts() # sanity check to see if replace function worked. Expect count to be around 304
```

# We are seeing a new miRNA list of 305 unique miRNAs. I expected 304 based on the outliers list, which includes 17 unique miRNAs. Most likely, one of these outlier miRNAs are already in the original list, so this would just add to an already existing count and not a new unique one.

```python
# Next, we need to extract the miRNAs from the miRBase and targetscan dataframes

mb_mirnas = set(df_mirna['mature_name'].unique())
ts_mirnas = set(df_targetscan['miRNA'].unique())
```

```python
# Check lengths of each set to compare

print(len(mb_mirnas & ts_mirnas)) # unique to each set
print(len(mb_mirnas)) # unique to miRBase only
print(len(ts_mirnas)) # unique to targetscan only
```

## Above check shows us what we knew before, that miRBase has 2656 unique miRNAs and targetscan 305. It also gives us a new insight, namely that the two sets share 296 unique miRNAs

```python
# inspect mismatches

sorted(list(ts_mirnas - mb_mirnas))
```

## Nine miRNAs in targetscan do not figure in the miRBAse list of miRNas. All but one of these miRNAs are named with mature -3p or -5p suffixes, which explains that. The last one, -325-3p does not exist in miRBase, however, it does as -325. This miRNA is a precursor and gives the name for the only registered mature miRNA, which comes off the 5-end of that precursor. -3p is mentioned in some papers, and has the sequence "UUUAUUGAGGACCUCCUAUCAA". This corresponds with a sequence section at further down towards the 3'end of the -325 precursor

```python
# hsa-miR-325-3p is not denoted in miRBase, however, hsa-miR-325 is. Checking if the theoretical seed and sequence for 325-3p exist 
# in the miRBase list in case it has another name

df_mirna[
    (df_mirna["seed"] == "UUUAUUG") &
    (df_mirna["sequence"] == "UUUAUUGAGGACCUCCUAUCAA")
]

# it does not exist
```

```python
# We are going to stick to a conservative approach when merging the dataframes. The 8 miRNAs from targetscan that do not show up in the shared
# list will be excluded. We will keep 325-3p, due to its strong prevalence in the targetscan data. The other 8 are not included simply because
# In this project, we do not want to over-complicate things by including miRNAs where the mature version is uncertain. At least for now.
# We remove 8 selected miRNAs later. First, we merge

df_merged = df_targetscan.merge(
    df_mirna,
    left_on='miRNA',
    right_on='mature_name',
    how='left',
    validate='many_to_one'
)
```

```python
# Checking overlap count, i.e. how many targetscan rows of miRNAs match the mature names from miRBase df

df_merged["mature_name"].isna().value_counts()

#This tells us that 8267 miRNA rows match, i.e. we are seeing many identical miRNAs from targetscan match a single miRNA mature name from miRBase df 
```

```python
# Let us confirm that we indeed have 296 unique miRNAs that are shared between both dataframes

df_merged['mature_name'].nunique()
```

```python
# Aggregating targetscan data to each miRNA to create the labels describing miRNA interaction properties
# We include mean and min score++ values, and number of interactions 

df_labels = (
    df_targetscan
    .groupby('miRNA', as_index = False) # keep miRNA column as regular column (not index)
    .agg(
        mean_context_score = ('context_score', 'mean'), # "mean" â†’ apply Series.mean() 
        min_context_score = ('context_score', 'min'), # "min" â†’ apply Series.min()
        n_targets = ('context_score', 'count') # "count" â†’ apply Series.count()
    )
)
```

```python
print(df_labels)

# mean score = average context score
# min score = strongest predicted context score
# n targets = number of targets per miRNA
```

```python
df_labels.describe()
```

```python
# We already see an outlier in df_labels (min score for min_context_score is below -1, which is outside the range (0 to -1))
# Let's QC this columns

df_labels.sort_values('min_context_score').head(10)


```

```python
# QC of labels dataframe

print(df_labels['min_context_score'].dtype)
print(df_labels['mean_context_score'].dtype)
print(df_labels[['mean_context_score', 'min_context_score']].isna().sum())
```

```python
# Let us remove 8 selected miRNAs from df_labels as discussed above

precursor_only = ['hsa-miR-137',
 'hsa-miR-147b',
 'hsa-miR-190b',
 'hsa-miR-217',
 'hsa-miR-320a',
 'hsa-miR-375',
 'hsa-miR-520b',
 'hsa-miR-520e'
 ]

df_labels = df_labels[~df_labels['miRNA'].isin(precursor_only)]
print(df_labels)

# Confirm the expected 297 rows of unique miRNAs

print(f'There are {df_labels['miRNA'].nunique()} unique miRNAs in df_labels')
```

```python
# COnfirm presence of specially selected miRNA, hsa-miR-325-3p

'hsa-miR-325-3p' in df_labels['miRNA'].values
```

```python
# Let's merge df_labels with selected features from df_mirna
# First, select features

selected_mirna_features = [
    'mature_name',
    'sequence',
    'length',
    'GC_content',
    'seed',
    'seed_GC_content',
    'seed_group_size'
]

df_mirna_features = df_mirna[selected_mirna_features]

df_mirna_features.head()
```

```python
# Merging miRNA features with labels

df_final = df_labels.merge( # df_labels define main row we want to keep 
    df_mirna_features, # this is the table we want to merge with
    left_on='miRNA', # the column identifier we match on 
    right_on='mature_name', # the equivalent colum to 'miRNA'
    how='left', # keep all rows from df_labels (important), add miRNA feature columns where there is a match, and if no match then NaN 
    validate='one_to_one' # safety check to control that each miRNA in df_label must match at most one in df_mirna_feature (protect from silent duplication)
)

df_final.head()
```

```python
# Some QC

print(df_final.shape)
print(df_final.describe())
```

```python
df_final = df_final.drop(columns=['mature_name'])
df_final.columns
```
